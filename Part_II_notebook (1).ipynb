````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
data = pd.read_csv('prosperLoanData.csv')
````

````python
Income_classes = ['$0', '$1-24,999', '$25,000-49,999', '$50,000-74,999', '$75,000-99,999', '$100,000+', 'Not employed', 
                'Not displayed']

Incomerange = pd.api.types.CategoricalDtype(ordered=True, categories=Income_classes)
data['IncomeRange'] = data['IncomeRange'].astype(Incomerange);
````

## (How Loan Original Amount is affected by Income range and Employment status)

> Loan original amount is the amount of loan gotten by a borrower. The highest loans where given to people in the 100,000 and above income range who were employed.

````python
#multivariate explorations
#check how LoanOriginalamount varies between Incomerange and Employment status
a_data = data[data["IncomeRange"].isin(['$25,000-49,999','$50,000-74,999', '$100,000+', '$75,000-99,999',  '$1-24,999'])]
plt.figure(figsize=(10, 6));
sb.barplot(x='EmploymentStatus', y='LoanOriginalAmount', hue='IncomeRange', data=a_data);
plt.xticks(rotation = 90);
plt.title('How INCOME RANGE AND EMPLOYMENT STATUS AFFECT LOAN ORIGINAL AMOUNT');
plt.ylabel('Loan Original Amount ($)');
plt.xlabel('Employment Status');
````

## (How High Prosper Scores and Employment Stataus affects Loan Original Amount)

> I noticed that borrowers who were employed and had high prosper scores got the biggest loans around 0 - 16,000 dollars.

````python
#check how higher prosper scores and employment status affects loan original amount
p_data = data[data["ProsperScore"].isin([5.0, 6.0, 7.0, 8.0, 9.0, 10.0, 11.0])]
plt.figure(figsize=(10, 8));
sb.barplot(x='EmploymentStatus', y='LoanOriginalAmount', hue='ProsperScore', data=p_data, palette = 'viridis');
plt.xticks(rotation = 90);
plt.title('HOW PROSPER SCORE AND EMPLOYMENT STATUS AFFECTS LOAN ORIGINAL AMOUNT');
plt.ylabel('Loan Original Amount ($)');
plt.xlabel('Employment Status');
````

Borrowers who were employed but had lower prosper scores got lower loans that their counterparts with higher scores. Getting loans in the region of 0 - 8000 dollars which is 6000 dollars lower in range.

````python
#check how lower prosper scores and employment status affects loan original amount
p_data = data[data["ProsperScore"].isin([1.0, 2.0, 3.0, 4.0])]
plt.figure(figsize=(10, 8));
sb.barplot(x='EmploymentStatus', y='LoanOriginalAmount', hue='ProsperScore', data=p_data, palette = 'viridis');
plt.xticks(rotation = 90);
plt.title('HOW PROSPER SCORE AND EMPLOYMENT STATUS AFFECTS LOAN ORIGINAL AMOUNT');
plt.ylabel('Loan Original Amount ($)');
plt.xlabel('Employment Status');
````

## (How Income Range and Employment Status Affects BorrowerAPR)

I noticed from the plot below that borrowers within the lowest income rangers usually got higher interest on their loans regardless of employment status as their loans are considered more risky.

````python
#check how borrowerapr changes according to income range and employment status
a_data = data[data["IncomeRange"].isin(['$25,000-49,999','$50,000-74,999', '$100,000+', '$75,000-99,999',  '$1-24,999'])]
plt.figure(figsize=(10, 8));
sb.barplot(x='EmploymentStatus', y='BorrowerAPR', hue='IncomeRange', data=a_data);
plt.xticks(rotation = 90);
plt.title('HOW INCOME RANGE AND EMPLOYMENT STATUS AFFECT BORROWER APR');
plt.ylabel('Borrower APR (%)');
plt.xlabel('Employment Status');
````
